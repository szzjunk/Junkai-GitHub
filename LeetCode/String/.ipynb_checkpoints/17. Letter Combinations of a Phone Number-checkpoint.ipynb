```python
def letterCombinations(digits):
    dictic={'2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl', '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'}
    output_list=[]
    nums_list=str(digits)
    if len(digits)==0:
        return []
    if len(digits)==1:
        return list(dictic[nums_list[0]])
    previous=dictic[nums_list[0]]
    for i in range(1,len(nums_list)):
        current = dictic[nums_list[i]]
        previous = [a + b for a in previous for b in current]
    return previous
```
